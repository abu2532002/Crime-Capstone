```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
csv_file_path = r'C:\Users\Admin\OneDrive\Documents\2011-IndiaStateDist-0000.csv'
df = pd.read_csv(csv_file_path)

# Group by state and sum the population for each state
state_population = df.groupby('State')['TOT_P'].sum().reset_index()

# Display the calculated population for each state
print(state_population)

# Plotting a bar chart for better visualization
plt.figure(figsize=(12, 6))
plt.bar(state_population['State'], state_population['TOT_P'])
plt.xlabel('State')
plt.ylabel('TOT_P')
plt.title('Population of Each State (2011)')
plt.xticks(rotation=90)
plt.show()
```

```python
import pandas as pd

# Load your CSV file
csv_file_path = r'C:\Users\Admin\OneDrive\Documents\2011-IndiaStateDist-0000.csv'
df = pd.read_csv(csv_file_path)

# Calculate literacy rate
df['LITERACY_RATE'] = (df['P_LIT'] / df['TOT_P']) * 100

# Group by state and calculate the average literacy rate for each state
state_literacy_rate = df.groupby('Name')['LITERACY_RATE'].mean().reset_index()

# Display the calculated literacy rate for each state
print(state_literacy_rate)
```
