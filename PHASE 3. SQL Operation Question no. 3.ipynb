```python
df = {
    'STATE_UT': ['UTTAR_PRADESH'],
    'DISTRICT': ['MUZAFFARNAGAR'],
    'MURDER': [287],
    'YEAR': [2002]
}

# Create a DataFrame from the result_data
result_dataframe = pd.DataFrame(df)

# Display the result DataFrame
print(result_dataframe)

# Group by STATE_UT and DISTRICT, count the number of unique years each DISTRICT appears
DISTRICT_counts = result_dataframe.groupby(['STATE_UT', 'DISTRICT']).size().reset_index(name='years_appeared')

# Filter districts that appear 3 or more years
filtered_DISTRICT = DISTRICT_counts[DISTRICT_counts['years_appeared'] >= 3]

# Merge with the original data to get the details
result_filtered = pd.merge(filtered_DISTRICT, result_dataframe, on=['STATE_UT', 'DISTRICT'])

# Display the result in descending order
result_filtered = result_filtered.sort_values(by=['years_appeared', 'STATE_UT', 'DISTRICT', 'YEAR'], ascending=[False, True, True, False])
print(result_filtered.reset_index(drop=True))
```

```python
import pandas as pd

df = {
    'STATE_UT': ['UTTAR_PRADESH', 'UTTAR_PRADESH', 'UTTAR_PRADESH'],
    'DISTRICT': ['MUZAFFARNAGAR', 'MUZAFFARNAGAR', 'MUZAFFARNAGAR'],
    'MURDER': [287, 253, 207],
    'YEAR': [2002, 2004, 2005]

}

# Create a DataFrame from the result_data
result_dataframe = pd.DataFrame(df)

# Display the result DataFrame
print(result_dataframe)

# Group by STATE_UT and DISTRICT, count the number of unique years each DISTRICT appears
DISTRICT_counts = result_dataframe.groupby(['STATE_UT', 'DISTRICT']).size().reset_index(name='years_appeared')

# Filter districts that appear 3 or more years
filtered_DISTRICT = DISTRICT_counts[DISTRICT_counts['years_appeared'] >= 3]

# Merge with the original data to get the details
result_filtered = pd.merge(filtered_DISTRICT, result_dataframe, on=['STATE_UT', 'DISTRICT'])

# Display the result in descending order
result_filtered = result_filtered.sort_values(by=['years_appeared', 'STATE_UT', 'DISTRICT', 'YEAR'], ascending=[False, True, True, False])
print(result_filtered.reset_index(drop=True))
```
