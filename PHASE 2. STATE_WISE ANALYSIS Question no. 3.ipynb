```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('03_District_wise_crimes_committed_against_children_2001_2012.csv',
                 usecols=['STATE/UT Name', 'Year', 'Total'])
print(df)
```

```python
df['Year'] = pd.to_datetime(df['Year'])
```

```python
Yearly_Total_Crime = df.groupby(df['Year'].dt.year)['Total'].sum().reset_index()

# Plotting the year-on-year total crime rate
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Total', data=Yearly_Total_Crime, marker='o')
plt.title('Year-on-Year Total Crime Rate')
plt.xlabel('Year')
plt.ylabel('Total Crime Rate')
plt.show()
```
