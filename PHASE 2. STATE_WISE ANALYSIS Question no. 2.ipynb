```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv('2011-IndiaStateDist-0000.csv',
                 usecols=['STATE/UT Name','P_LIT'])
print(df1)
```

```python
df2 = pd.read_csv('03_District_wise_crimes_committed_against_children_2001_2012.csv',
                 usecols=['STATE/UT Name', 'Murder', 'Rape', 'Kidnapping and Abduction', 'Foeticide', 'Abetment of suicide', 'Exposure and abandonment', 'Procuration of minor girls', 'Buying of girls for prostitution', 'Selling of girls for prostitution', 'Prohibition of child marriage act', 'Other Crimes'])
print(df2)
```

```python
df = pd.concat([df1, df2], axis=1)
print(df)
```

```python
# Assuming df1 and df2 are your two DataFrames

# Check and handle any issues with df1
df1['STATE/UT Name'] = df1['STATE/UT Name'].astype(str)
value_counts_df1 = df1['STATE/UT Name'].value_counts()

# Check and handle any issues with df2
df2['STATE/UT Name'] = df2['STATE/UT Name'].astype(str)
value_counts_df2 = df2['STATE/UT Name'].value_counts()

# Display the value counts for both DataFrames
print("Value Counts for df1:")
print(value_counts_df1)

print("\nValue Counts for df2:")
print(value_counts_df2)
```

```python
# Assuming 'STATE/UT Name' is the common column in both DataFrames
merged_data = pd.merge(df1, df2, on='STATE/UT Name', how='inner')

# Display the first few rows of the merged DataFrame
print("\nMerged Data:")
print(merged_data.head())
```

```python
# List of crime-related columns
crime_columns = ['Murder', 'Rape', 'Kidnapping and Abduction', 'Foeticide', 'Abetment of suicide', 'Exposure and abandonment', 'Procuration of minor girls', 'Buying of girls for prostitution', 'Selling of girls for prostitution', 'Prohibition of child marriage act', 'Other Crimes']

# Print unique values for each crime-related column
for column in crime_columns:
    print(f"Unique values in {column}: {df[column].unique()}")

# Assuming 'melted_df' is your melted DataFrame
melted_df = pd.melt(merged_data, id_vars=['STATE/UT Name', 'P_LIT'], var_name='Crime_Type', value_name='Count')

# Display the first few rows of the modified melted DataFrame
print("\nModified Melted Data:")
print(melted_df.head())
```

```python
# Assuming 'melted_df' is your melted DataFrame
melted_df = pd.melt(merged_data, id_vars=['STATE/UT Name', 'P_LIT'], var_name='Crime_Type', value_name='Count')

# Display the first few rows of the modified melted DataFrame
print("\nModified Melted Data:")
print(melted_df.head())
```

```python
# Assuming 'melted_df' is your melted DataFrame
melted_df['P_LIT'] = pd.to_numeric(melted_df['P_LIT'], errors='coerce')  # Ensure 'P_LIT' is numeric

# Group by 'STATE/UT Name' and 'Crime_Type', then calculate the mean for both 'P_LIT' and 'Count'
grouped_data = melted_df.groupby(['STATE/UT Name', 'Crime_Type']).agg({'P_LIT': 'mean', 'Count': 'sum'}).reset_index()

# Print the grouped data
print("\nGrouped Data:")
print(grouped_data)

crime_type = 'Murder'
sns.scatterplot(x='P_LIT', y='Count', hue='STATE/UT Name', data=grouped_data[grouped_data['Crime_Type'] == crime_type])
plt.title(f'{crime_type} vs P_LIT by STATE/UT Name')
plt.show()

# Correlation analysis
correlation = grouped_data['P_LIT'].corr(grouped_data['Count'])
print(f'Correlation between P_LIT and Count: {correlation}')
```
